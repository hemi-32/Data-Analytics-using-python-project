```python
mark1=int(input("Enter the marks 1:"))
mark2=int(input("Enter the marks 2:"))
mark3=int(input("Enter the marks 3:"))
Total=mark1+mark2+mark3
Average=Total/3
print("Total marks:",Total)
print("Average marks:",Average)
if Average>=40:
    print("Pass")
else:
    print("Fail")

```

```python
marks={
    "Tamil":90,
    "english":87,
    "maths":98,
    "science":94,
    "social":94
    }
total=0
for subject,mark in marks.items():
    print(subject,":",mark)
    total +=mark
print("total marks:",total)
if total>=450:
  print("Grade A")
elif total>=400:
  print("Grade B")
elif(total>=350):
  print("Grade C")
else:
  print("Grade D")


```
