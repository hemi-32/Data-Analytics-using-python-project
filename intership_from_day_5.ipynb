```python
import pandas as pd

employee={
    "Emp_ID":[101,102,103,104,105],
    "Name":["mathu","prabha","viji","santhiya","karthika"],
    "Department":["medical","nursing","IT","HR","doctor"],
    "Salary":[50000,60000,70000,80000,90000],
    "Experience":[2,3,4,5,6]
}
df=pd.DataFrame(employee)
print("Employee Dataset")
print(df)

print("\nAverage Salary")
print(df["Salary"].mean())

print("\nHighest Salary")
print(df["Salary"].max())

print("\nEmployee in HR")
print(df[df["Department"]=="HR"])

print("\nEmployees with experience greater than 5 years")
print(df[df["Experience"]>5])

print("\nEmployee Count by Department")
print(df["Department"].value_counts())





```

# **TASK**-clean a messy sales Dataset

```python
data={
    "Employee_id":[101,102,103,104,105],
    "Name":["mathu","prabha","viji","santhiya","karthika"],
    "Age":[25,26,29,40,45],
    "Salary":[50000,60000,70000,80000,90000],
}
df=pd.DataFrame(data)
print(df)

print(df.isnull())

df=(df.drop_duplicates())
print(df)


df=df.dropna(subset=["Salary"])
print(df)


df=df.rename(columns={"Employee_ID":"Emp_ID"})
print(df)

```

# **data filtering&transformation**

```python
data={
    "Region":["North","South","East","West","South","North"],
    "Product":["laptop","mobile","laptop","Tablet","Mobile","Tablet"],
    "Sales":[50000,30000,45000,250000,2000,35000]

}
df=pd.DataFrame(data)
print(df)
```

```python
south_sales=df[df["Region"]=="South"]
print(south_sales)
```

```python
high_sales=df[df["Sales"]>30000]
print(high_sales)
```

```python
sorted_df=df.sort_values(by="Sales")

print(sorted_df)
```

**ground by-group similar records together**

```python
region_sales=df.groupby("Region")["Sales"].sum()
print(region_sales)
```

```python
Product_sales=df.groupby("Product")["Sales"].sum()
print(Product_sales)
```

```python
print(df["Sales"].sum())
```

```python
print(df["Sales"].mean())
```

```python
print(df["Sales"].max())
```

```python
print(df["Sales"].min())
```

```python
print(df["Sales"].count())
```

```python
result=df.groupby("Region")["Sales"].agg(["sum","mean","max","min","count"])
print(result)
```

**task**-Region-wise Sales Report

```python
df_active=df[df['Sales']>0]
df_stored=df_active.sort_values(by=['Region','Sales'],ascending=[True,False])
report=df_stored.groupby('Region')['Sales'].agg(['sum','mean','max','min','count'])
report.columns=['Total Sales','Average Sales','Highest Sales','Lowest Sales','Number of Sales']
print(report)
```

# **task-line chart pie chart bar chat**

```python
import matplotlib.pyplot as plt

month=["jan","feb","mar","apr","may","jun"]
sales=[10000,12000,15000,13000,18000,20000]

# Line chart
plt.figure(figsize=(5,3))
plt.plot(month,sales,marker="o")
plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()

# Bar chart
plt.figure(figsize=(5,3))
plt.bar(month,sales,color="orange")
plt.title("Sales Comparison")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()

# Pie chart
expenses=[5000,3000,2000]
categories=["Food","Travel","shopping"]

plt.figure(figsize=(5,5))
plt.pie(expenses,labels=categories,autopct="%1.1f%%",colors=['red','blue','pink'])
plt.title("Expenses Distribution")
plt.tight_layout()
plt.show()
```
