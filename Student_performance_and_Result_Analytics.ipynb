```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

uploaded=files.upload()

df=pd.read_csv("student-dataset.csv")

print(df)
```

```python
print(df.shape)
```

```python
df=df.drop_duplicates()
df['Total_Score'] = df[['english.grade', 'math.grade', 'sciences.grade', 'language.grade',
                        'portfolio.rating', 'coverletter.rating', 'refletter.rating']].sum(axis=1)
print(df)
```

```python
df.fillna(0,inplace=True)
print(df)
```

```python
print(df.isnull().sum())
```

```python
total_scores=df['Total_Score'].sum()
print("Total_scores:",total_scores)
```

```python
avg_score_by_nationality = df.groupby('nationality')['Total_Score'].mean().reset_index()
print("Average Total Score by Nationality:")
display(avg_score_by_nationality)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Total_Score'],bins=10,edgecolor='black',linewidth=1.2,color='skyblue')
plt.xlabel('Students')
plt.ylabel('Total Score')
plt.title('Total Score Distribution')
plt.show()
```

```python
top10=avg_score_by_nationality.head(10)

plt.plot(top10['nationality'],top10['Total_Score'],marker='o')
plt.xticks(rotation=30)
plt.show()
```

```python
df['gender'].value_counts().plot(kind='pie',autopct='%1.1f%%',startangle=90,figsize=(6,6))
plt.ylabel('')
plt.title('Gender Distribution')
plt.show()
```
